The mean shows the average value of numerical columns like Sales, Quantity, Discount, and Profit.
The median represents the middle value and helps identify skewness in the data. 
The mode indicates the most frequently occurring value, such as Discount being 0 in most cases.
The standard deviation measures the variability in the data. A high standard deviation indicates large variation,while a low value shows consistency.

```python
# create a bar chart 
import matplotlib.pyplot as plt

cat_cols = df.select_dtypes(include=['object'])

for col in cat_cols.columns:
    plt.figure(figsize=(8,4))
    df[col].value_counts().plot(kind='bar', color='red')
    plt.title(f'Bar Chart of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()
```
