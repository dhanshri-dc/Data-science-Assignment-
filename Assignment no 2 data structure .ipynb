Q1:Create a list of 5 integers. Perform and print the result of the following operations: 
append, extend, insert, remove, pop, clear, index, count, sort, and reverse.

```python
#Append(): It will adds an element at the end of the list

numbers = [1,2,3,4]
numbers.append(5)
print(numbers)
```

```python
#extend():It will adds elements from another list
a = [1,2]
b = [3,4]
a.extend(b)
print(a)
```

```python
#insert():insert an element at a specified index
nums = [10,20,30]
nums.insert(1,15)
print(nums)
```

```python
# Remove():It will removes the first occurence of specified values

nums = [10,20,"abc",30,40]
nums.remove(10)
print(nums)
```

```python
#pop():removes and returns an element (default last element will be remove)

data = [10,20,"Python",30,40.55]
data.pop()
print(data)
```

```python
#clear(): It will return empty list
lst = [1,2,3]
lst.clear()
print(lst)
```

```python
# index(): It will return the index of first occurenec of a value

nums = [10,20,30]
print(nums.index(20))
```

```python
# Count() : Counts how many times value appears in datastructure

list_1 = [10,50,"WEB",89,100,50]
print(list_1.count(50))
```

```python
# sort() : Sorts the data in ascending order(by default)
nums = [3,1,2,4,9,5]
nums.sort()
print(nums)
```

```python
# Reverse : value is reversed
nums =[1,2,3,4,5]
nums.reverse()
print(nums)
```

Q7.Given a number, check if it's positive, and if it is also even. If not positive, print if it's zero or negative.

```python
number = int(input("Enter a number: "))

if number > 0:
    if number % 2 == 0:
        print("The number is positive and even.")
    else:
        print("The number is positive but odd.")
elif number == 0:
    print("The number is zero.")
else:
    print("The number is negative.")
```

Q11.Create a small DataFrame manually with 10 rows and columns: 'Name', 'Age', 'City', and 'Salary'. Then:
- Use .info(), .describe()
- Select 'Name' and 'City' columns
- Drop 'City' column
- Fill any missing values in 'Salary' column with the mean
- Remove any duplicate rows


```python
import pandas as pd
import numpy as np

# Create DataFrame manually
data = {
    'Name': ['John', 'Alice', 'Bob', 'David', 'Emma',
             'Frank', 'Grace', 'Henry', 'Ivy', 'John'],
    'Age': [25, 30, 22, 35, 28,
            40, 27, 32, 29, 25],
    'City': ['New York', 'London', 'Paris', 'Tokyo', 'Berlin',
             'Sydney', 'Toronto', 'Dubai', 'Rome', 'New York'],
    'Salary': [50000, 60000, np.nan, 80000, 75000,
               90000, np.nan, 65000, 70000, 50000]
}

df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)
```

```python

df.info()
```

```python
df.describe()
```

```python
selected_columns = df[['Name', 'City']]
print("\nSelected Columns (Name & City):")
print(selected_columns)
```

```python
df = df.drop(columns=['City'])
print("\nAfter Dropping 'City' Column:")
print(df)
```

```python
mean_salary = df['Salary'].mean()
df['Salary'].fillna(mean_salary, inplace=True)

print("\nAfter Filling Missing Salary with Mean:")
print(df)
```

```python
df = df.drop_duplicates()

print("\nAfter Removing Duplicates:")
print(df)
```
