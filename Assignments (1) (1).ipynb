```python
#Write a Python program that checks whether a given number is prime or not.
#A prime number is a natural number greater than 1 that has no positive divisors other than 1 and itself.
##Answer : A prime number is natural number greater than 1 that has no positive divisionrs other than 1 and itself.
#This program chechs whether the given number is prime or not 
#Code:
def is_prime(number):
    if number <= 1:
        return False
    if number == 2:       
        return True
    if number % 2 == 0:
        return False
```

```python
num = int(input("Enter a number: "))

if is_prime(num):
    print(f"{num} is a prime number.")
else:
    
    print(f"{num} is not a prime number.")
```

#= Exercise 4: Hospital Billing System 

# A hospital charges its patients based on the following:
•	Room charges per day: ₹2,000
•	Doctor consultation fee (flat): ₹1,500
•	Lab test charges: Based on the number of tests taken. Each test costs ₹300
•	Medicine charges: Total cost of medicines provided
•	Discount: If the total bill (before discount) exceeds ₹10,000, a 10% discount is applied


```python
room_charge_per_day = 2000
doctor_fee = 1500
lab_test_charge = 300

days = 4
lab_tests = 3
medicine_charge = 2400

room_total = room_charge_per_day * days
lab_total = lab_test_charge * lab_tests

total = room_total + doctor_fee + lab_total + medicine_charge

if total > 10000:
    discount = total * 0.10
    final_bill = total - discount
else:
    final_bill = total

print("Total before discount:", total)
print("Final Bill:", final_bill)
print("Discount:(10%)=1280")
```
