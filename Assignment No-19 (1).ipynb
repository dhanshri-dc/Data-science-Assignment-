```python
import pandas as pd
import numpy as np
import re
import nltk

from nltk.corpus import stopwords
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
```

```python
data = {
    "label": ["pos","pos","pos","neg"],
    "review": [
        "Amazing soundtrack I love it",
        "Great book very interesting",
        "Worst product ever",
        "Bad quality waste of money"
    ]
}

df = pd.DataFrame(data)
```

```python
df
```

```python

def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-zA-Z]', ' ', text)
    return text

sample_text = "Stuning even for the non-gamer: This sound track was beautiful! 123"
cleaned = clean_text(sample_text)
print(cleaned)
```

```python
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

def remove_stopwords(text):
    words = text.split()
    words = [w for w in words if w not in stop_words]
    return " ".join(words)
```

```python
df['review'] = df['review'].apply(clean_text)
df['review'] = df['review'].apply(remove_stopwords)
```

```python
df['label'] = df['label'].map({'pos':1, 'neg':0})
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    df['review'], df['label'], test_size=0.2, random_state=42
)
```

```python
vectorizer = TfidfVectorizer(max_features=5000)

X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression

# Drop rows with missing labels
df = df.dropna(subset=['label'])

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(
    df['review'], df['label'], test_size=0.2, random_state=42
)

# Vectorize text
vectorizer = TfidfVectorizer()
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)

# Train model
model = LogisticRegression(max_iter=500)
model.fit(X_train_vec, y_train)

print("Model trained successfully!")
```

# Sentiment Analysis of Reviews
## 1. Introduction

Sentiment Analysis is the process of determining the emotional tone behind a body of text. It helps understand whether a piece of text expresses a positive, negative, or neutral sentiment.

Problem Statement:
We aim to analyze customer reviews and classify them into positive or negative sentiments. This will help businesses understand customer opinions automatically and at scale.

# 2. Data Preprocessing

Before training the model, the text data needs to be cleaned and standardized:

Converted all text to lowercase to avoid case-sensitive differences.
Removed special characters, numbers, and punctuation to focus on meaningful words.
Removed missing values (NaN) to avoid errors during training.
Optionally, stopwords and stemming/lemmatization can be applied to improve performance.

This ensures that the input text is consistent and ready for vectorization

# 3. Exploratory Data Analysis (EDA)
Counted the number of positive vs negative reviews to check class balance.
Visualized the distribution of sentiments (bar chart/pie chart recommended).

This step helps understand the dataset and ensures that the model is not biased toward one class.

# 4. Model

We used TF-IDF (Term Frequency-Inverse Document Frequency) to convert text into numerical vectors. TF-IDF gives higher weight to important words and reduces the impact of common words.

Model Choice: Logistic Regression

Simple, fast, and effective for text classification.
Works well with high-dimensional sparse data from TF-IDF vectors.
