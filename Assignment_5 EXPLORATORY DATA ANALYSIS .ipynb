```python
## 3.Data Visualization:
# this histogram shows only LB (his),(boxplot)
import matplotlib.pyplot as plt

# Histogram for LB column
plt.hist(df['LB'], bins=10, color='skyblue', edgecolor='black')

plt.title('Distribution of LB')
plt.xlabel('LB Values')
plt.ylabel('Frequency')

plt.show()
```

```python
import matplotlib.pyplot as plt

plt.boxplot(df['LB'])

plt.title('Box Plot of LB')
plt.ylabel('LB Values')

plt.show()
```

```python
##this histogram shows the all values in one code line ##
df[['LB','AC','FM','UC']].hist(figsize=(10,8))
plt.show()
```

```python
import matplotlib.pyplot as plt

# Frequency of NSP categories
df['NSP'].value_counts().plot(kind='bar')

plt.title('Frequency of NSP Categories')
plt.xlabel('NSP Category')
plt.ylabel('Count')

plt.show()
```

```python
import matplotlib.pyplot as plt

df['NSP'].value_counts().plot(kind='pie', autopct='%1.1f%%')

plt.title('Distribution of NSP Categories')
plt.ylabel('')

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt 
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(),annot=True,cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.pairplot(df[['LB','AC','FM','UC','NSP']])
plt.show()
```

```python
# The correlation heatmap and scatter plots help in identify relationship between different
#fetal health parameter . some variables show moderate correlation ,which may help in predicting the fetal state (NSP)
```

```python
sns.pairplot(df[['LB','AC','FM','UC','NSP']],hue='NSP')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.violinplot(x='NSP', y='LB', data=df)
plt.title("Violin plot of LB vs NSP")
plt.show()
```

```python
## 4.Pattern Recognition and Insights:
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), annot=True,cmap='coolwarm')
plt.title("Correlation Between Variables")
plt.show()
```

```python
# look for trends or pattern 
plt.scatter(df['LB'], df['MLTV'])
plt.xlabel("Baseline Heart Rate (LB)")
plt.ylabel("Mean Long Term Variability(MLTV)")
plt.title("Trend Between LB and MLTV")
plt.show()
```

## 5 .Conclusion ##
The exploratory data analysis of the cardiotocography dataset provided useful insights into the relationships and distributions of various fetal health parameters.    
From the statistical summaries and visualizations, it was observed that several variables such as LB (baseline fetal heart rate), AC (accelerations), FM (fetal movements), and UC (uterine contractions) show variation across the dataset. Visualization techniques such as histograms, boxplots, scatter plots, correlation heatmaps, pair plots, and violin plots helped in understanding the distribution and relationships between variables.

The correlation analysis revealed that some variables have moderate relationships with each other, which may influence the fetal state represented by the NSP variable. These patterns help in identifying potential indicators of fetal health conditions.
